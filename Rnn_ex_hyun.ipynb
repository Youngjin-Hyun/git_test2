```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
from tensorflow.python.ops import rnn, rnn_cell
from tensorflow.python.ops.rnn_cell import BasicRNNCell, BasicLSTMCell

batch_size =16
mnist =input_data.read_data_sets('./tmp/data/',one_hot=True)
```

```python
n_input = 28
timesteps =28
n_classes =10
```

```python
num_hidden =64
num_layers =2
```

```python
x = tf.placeholder(tf.float32,[None,timesteps,n_input])
y = tf.placeholder(tf.float32,[None,n_classes])

weights = {
    'out':tf.Variable(tf.random_normal([num_hidden,n_classes]))
}
biases = {'out':tf.Variable(tf.random_normal([n_classes]))}
```

```python
def RNN(x, weights,biases):
    x =tf.unstack(x, timesteps,1)
    stack_rnn=[]
    for i in range(num_layers):
        cell = tf.nn.rnn_cell.BasicLSTMCell(num_hidden)
        cell = tf.nn.rnn_cell.DropoutWrapper(cell=cell,output_keep_prob=0.8)
        stack_rnn.append(cell)
    stacked_cell = tf.nn.rnn_cell.MultiRNNCell(stack_rnn,state_is_tuple=True)
    outputs, last_states = rnn.static_rnn(stacked_cell,x,dtype=tf.float32)
    return tf.matmul(outputs[-1],weights['out']+biases['out'])
logits =RNN(x, weights,biases)
print("logits",logits)
```

```python
### train and evaluation
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits,labels=y))
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(10):
        total_batch = int(mnist.train.num_examples/batch_size)
        average_cost = 0
        for i in range(total_batch):
            batch_xs, batch_ys = mnist.train.next_batch(batch_size)
            batch_xs = batch_xs.reshape((batch_size,timesteps,n_input))
            loss,_ = sess.run([cross_entropy, train_step], feed_dict={x:batch_xs, y:batch_ys})
            
            average_cost+=loss/total_batch
            print('Epoch %d,loss: %f' % (epoch,average_cost))
            pred = tf.nn.softmax(logits)
            correct_prediction = tf.equal(tf.argmax(pred,1),tf.argmax(y,1))
            accuracy =tf.reduce_mean(tf.cast(correct_prediction,"float"))
            mnist_test = mnist.test.images.reshape((-1,timesteps,n_input))
            print("Accuracy:", accuracy.eval({x:mnist_test,y:mnist.test.labels}))
            
```
